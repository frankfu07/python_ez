```python
# create a function to check missing data and generate report
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending=False)
    percent = (data.isnull().sum()/len(data)*100).sort_values(ascending = False)
    return pd.concat([total,percent], axis=1,keys=['Total', 'Percent']) # When axis=1, a DataFrame is returned
```

```python
missing_data(df)
```

```python
#check correlations
ext_data=df[['TARGET', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']]
ext_data_corrs = ext_data.corr()
ext_data_corrs
```

```python
#  heatmap
sns.heatmap(ext_data_corrs, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap');
# EXT_SOURCE featureshave negative correlations with the target, 
# indicating that as the value of the EXT_SOURCE increases, the client is more likely to repay the loan
```
