### 4. Basic Pandas concepts


#### 4.1 Pandas Series

A Series is the fundational data structure of Pandas. It contains two types of information: index and values. The default index is [0,1,2,3,...n]

```python
basic_series = pd.Series(["a","b","c"])
basic_series
```

To retrieve a certain value, specify the index.

```python
basic_series[0]
```

#### 4.2 Pandas DataFrame

A Pandas DataFrame is multiple Sereis combined together with a common index. The indiviudal Series are called columns.



```python
sample_df = pd.DataFrame({"Instructors": ["Ken", "Steven", "Xing"], "Heights":[80,70,68], "Locations":["ON", "BC", "QC"]})
sample_df
```

To select a single column from a dataframe, use the dataframe name along name like df['column_name']

```python
sample_df['Locations']
```

```python
sample_df[['Locations']]
```
